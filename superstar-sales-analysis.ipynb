```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python
df['Order Date'].min(), df['Order Date'].max()#Total time period
```

```python
df['Order Year']=df['Order Date'].dt.year
df['Order Month']=df['Order Date'].dt.month
```

```python
df.head()
```

```python
df[['Sales','Profit']].sum()#Total sales and profit
```

```python
df.groupby('Order Year')['Sales'].sum()
df.groupby('Category')['Sales'].sum()
df.groupby('Region')['Profit'].sum()
```

```python
top_sales_product=df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_sales_product
```

```python
Top_Profit_poduct=df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)
Top_Profit_poduct
```

```python
Top_Loss_poduct=df.groupby('Product Name')['Profit'].sum().sort_values().head(10)
Top_Loss_poduct
```

```python
# STEP 8: Discount vs Profit Analysis

# 8.1 Total profit by discount
discount_profit = df.groupby('Discount')['Profit'].sum()

# 8.2 Average profit by discount
discount_avg_profit = df.groupby('Discount')['Profit'].mean()

# 8.3 Discount ranges
df['Discount Range'] = pd.cut(
    df['Discount'],
    bins=[-0.01, 0.1, 0.3, 0.6, 1.0],
    labels=['No/Low', 'Medium', 'High', 'Very High']
)

# 8.4 Profit by discount range
discount_range_profit = df.groupby('Discount Range')['Profit'].sum()

discount_profit, discount_avg_profit, discount_range_profit

```

```python
# STEP 9: Customer & Segment Analysis

# 9.1 Segment-wise Sales
segment_sales = df.groupby('Segment')['Sales'].sum()

# 9.2 Segment-wise Profit
segment_profit = df.groupby('Segment')['Profit'].sum()

# 9.3 Segment-wise number of orders
segment_orders = df.groupby('Segment')['Order ID'].nunique()

# 9.4 Combine all metrics
segment_summary = pd.DataFrame({
    'Total Sales': segment_sales,
    'Total Profit': segment_profit,
    'Total Orders': segment_orders
})

segment_summary
```

```python
# STEP 10: Final Business Insights

# Overall KPIs
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
avg_discount = df['Discount'].mean()

summary = {
    'Total Sales': total_sales,
    'Total Profit': total_profit,
    'Total Orders': total_orders,
    'Average Discount': avg_discount
}

summary
```
